# The Data Engineer's Salary Saga

In 2024, the landscape of data engineering witnessed dynamic shifts, reflected notably in the compensation patterns of professionals in the field. A comprehensive analysis of data engineer salaries offers valuable insights into the evolving trends, providing crucial information for both job seekers and employers navigating this rapidly changing terrain.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

The dataset for 2024 reveals a diverse range of salary distributions, reflecting the varied experience levels and employment types within the data engineering domain. From entry-level positions to seasoned senior engineers, the spectrum of salaries showcases the importance of experience and expertise in shaping compensation packages. Notably, the median salary for data engineers has seen a steady rise, indicating the increasing demand for skilled professionals adept at handling complex data infrastructures and analytics frameworks.

```python
df = pd.read_csv("/kaggle/input/data-engineer-salary-in-2024/salaries (2).csv")
```

Moreover, the geographical distribution of salaries unveils intriguing disparities, with certain regions emerging as hotspots for lucrative compensation packages. Factors such as employee residence, company location, and remote work opportunities contribute significantly to salary differentials, highlighting the influence of globalization and remote work trends on remuneration structures.

```python
# Set the style for the plots
plt.style.use('ggplot')
```

Additionally, company size plays a pivotal role in shaping data engineer salaries, with larger enterprises typically offering higher compensation packages compared to their smaller counterparts. This trend underscores the strategic importance of data engineering in driving business growth and innovation, prompting larger organizations to invest substantially in attracting and retaining top talent in the field.

```python
# Data Visualization 1: Salary Distribution
plt.figure(figsize=(10, 6))
plt.hist(df['salary_in_usd'], bins=30, color='orange', alpha=0.7)
plt.title('Salary Distribution in 2024')
plt.xlabel('Salary (USD)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# Data Visualization 2: Experience Level Distribution
plt.figure(figsize=(8, 6))
df['experience_level'].value_counts().plot(kind='bar', color='orange', alpha=0.7)
plt.title('Experience Level Distribution')
plt.xlabel('Experience Level')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.show()
```

```python
# Data Visualization 3: Employment Type Distribution
plt.figure(figsize=(8, 6))
df['employment_type'].value_counts().plot(kind='bar', color='orange', alpha=0.7)
plt.title('Employment Type Distribution')
plt.xlabel('Employment Type')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.show()
```

```python
# Data Visualization 4: Salary vs. Experience Level
plt.figure(figsize=(10, 6))
sns.boxplot(x='experience_level', y='salary_in_usd', data=df, palette='Oranges')
plt.title('Salary vs. Experience Level')
plt.xlabel('Experience Level')
plt.ylabel('Salary (USD)')
plt.show()
```

```python
# Data Visualization 5: Salary vs. Employment Type
plt.figure(figsize=(10, 6))
sns.boxplot(x='employment_type', y='salary_in_usd', data=df, palette='Oranges')
plt.title('Salary vs. Employment Type')
plt.xlabel('Employment Type')
plt.ylabel('Salary (USD)')
plt.show()
```

```python
# Data Visualization 6: Remote Work Ratio Distribution
plt.figure(figsize=(8, 6))
df['remote_ratio'].value_counts().plot(kind='bar', color='orange', alpha=0.7)
plt.title('Remote Work Ratio Distribution')
plt.xlabel('Remote Work Ratio')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.show()
```

```python
# Data Visualization 7: Salary vs. Remote Work Ratio
plt.figure(figsize=(10, 6))
plt.scatter(df['remote_ratio'], df['salary_in_usd'], color='orange', alpha=0.5)
plt.title('Salary vs. Remote Work Ratio')
plt.xlabel('Remote Work Ratio')
plt.ylabel('Salary (USD)')
plt.grid(True)
plt.show()
```

```python
# Data Visualization 8: Company Size Distribution
plt.figure(figsize=(8, 6))
df['company_size'].value_counts().plot(kind='bar', color='orange', alpha=0.7)
plt.title('Company Size Distribution')
plt.xlabel('Company Size')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.show()
```

```python
# Data Visualization 9: Salary vs. Company Size
plt.figure(figsize=(10, 6))
sns.boxplot(x='company_size', y='salary_in_usd', data=df, palette='Oranges')
plt.title('Salary vs. Company Size')
plt.xlabel('Company Size')
plt.ylabel('Salary (USD)')
plt.show()
```

```python
# Data Visualization 10: Top 10 Countries with Highest Average Salary
top_countries_salary = df.groupby('employee_residence')['salary_in_usd'].mean().nlargest(10)
plt.figure(figsize=(10, 6))
top_countries_salary.plot(kind='bar', color='orange', alpha=0.7)
plt.title('Top 10 Countries with Highest Average Salary')
plt.xlabel('Country')
plt.ylabel('Average Salary (USD)')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

In conclusion, the analysis of data engineer salaries in 2024 underscores the dynamic nature of the data engineering landscape, characterized by evolving compensation patterns influenced by factors such as experience, geography, employment type, and company size. As organizations continue to prioritize data-driven decision-making and digital transformation initiatives, the demand for skilled data engineers is expected to remain robust, driving further evolution in salary trends and employment dynamics within the field.
